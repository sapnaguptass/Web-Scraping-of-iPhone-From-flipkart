```python
file_name = "product2.csv" #product.csv will be created where your code is located
file = open(file_name,"w",encoding="utf-8") #open file in write mode
```

```python
#Desing header of your csv file
header = "Product Name , Price ,FinalPrice,Prod Rating, RAM , Display, Camera,Warranty \n"
file.write(header)#write your header in file
```

```python

for container in containers:
    name=container.findAll("div",{"class":"_3wU53n"})
    prodName=name[0].text
    #print(name[0].text)
    rating=container.findAll("div",{"class":"hGSR34"})
    prodRating=rating[0].text
    #print(rating[0].text)
    information=container.findAll("li",{"class":"tVe95H"})
    RAM = information[0].text
    Display=information[1].text
    Camera =information[2].text
    Warranty=information[4].text
    
    
    prodInformation=information[0].text
    #print(information[0].text)
    price=container.findAll("div",{"class":"col col-5-12 _2o7WAb"})
    prodPrice=price[0].text
    #print(price[0].text.strip())
    price_only =''.join(prodPrice.split(','))#first split the price with comma then join it
    
    rm_rupees=price_only.split('₹')
    add_rs_price='Rs ' + rm_rupees[1]
    final_price=add_rs_price
    #file.write(prodName + ',' + prodRating + ',' + prodPrice + ',' + prodInformation +'\n')
    file.write(prodName + ',' +final_price + ',' + prodRating + ',' + RAM + ',' + Display+',' + Camera+',' + Warranty+'\n')

file.close()
    
    
```
